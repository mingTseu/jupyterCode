# K近邻算法

## 读入数据

```python
import pandas as pd 
import numpy as np

bc_data=pd.read_csv("bc_data.csv",header=0)
bc_data.head()
#默认5行
```

## 数据理解

```python
bc_data.shape
```

```python
bc_data.columns
```

```python
#数据的统计性描述
bc_data.describe() 
```

## 数据准备

```python
#axis=1 表示计算前后行数不变，以行为单位逐行删除
data=bc_data.drop(["id"],axis=1)
data.head()
```

```python
#去掉因变量，形成X
X_data=data.drop(["diagnosis"],axis=1)
X_data.head()
```

```python
#将y降维值一维向量
y_data=np.ravel(data[["diagnosis"]])
#Fancy indexing，应用无规则下表切分[[1,3,5]]
y_data[0:6]
```

## 切分训练集、测试集

```python
from sklearn.model_selection import train_test_split
#使用train_test_split函数
X_trainSet,X_testSet,y_trainSet,y_testSet = train_test_split(X_data,y_data,random_state=1)
print(X_trainSet.shape)
```

## 算法选择及超参设置

```python
from sklearn.neighbors import KNeighborsClassifier
myModel = KNeighborsClassifier(algorithm='kd_tree')
#ball_tree,kd_tree,brute,自动autob
#默认k=5
```

```python
myModel.fit(X_trainSet,y_trainSet)
#n_jobs=1几个cpu、metric哪种距离   minkowski p是个参数  leaf_size最外围
```

## 利用模型预测

```python
y_predicSet=myModel.predict(X_testSet)
print(y_predicSet)
```

```python
print(y_testSet)
```

## 模型评价

```python
from sklearn.metrics import accuracy_score
print("准确率为：",accuracy_score(y_testSet,y_predicSet))
```

```python
NumOfK=range(1,23)
KNNs = [KNeighborsClassifier(n_neighbors=i) for i in NumOfK]
scores = [KNNs[i].fit(X_trainSet,y_trainSet).score(X_testSet,y_testSet) for i in range(len(KNNs))]
scores
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']="simHei"
plt.plot(NumOfK,scores)
plt.title('癌症预测KNN模型中k对准确率的影响')
plt.xlabel("K值")
plt.ylabel("准确率")
plt.show()
```
